## **🟢 Part 1: Loan Eligibility & Interest Rate Calculation [4 marks]**  

📌 **Requirement:**  
SecureBank only approves loans if the **customer meets all eligibility criteria**. The system should:  
✔ **Check if the customer is employed.**  
✔ **Verify if the income is at least PKR 50,000.**  
✔ **Check credit score:**  
   - **750+ → 5% Interest Rate**  
   - **650 - 749 → 8% Interest Rate**  
   - **Below 650 → Loan Rejected**  
✔ **If the applicant is unemployed, the loan is rejected immediately.**  

### **Implementation Steps:**  
🔹 **Step 1:** Structure the `if-elif-else` statements for this logic.  
🔹 **Step 2:** Implement a Python program using **ipywidgets** for interactive user input (income, credit score, employment status) to decide whether the loan is approved or rejected.

---

```python
import ipywidgets as widgets
from ipywidgets import interact

income_slider = widgets.IntSlider(value=50000, min=0, max=200000, step=1000, description="Annual Income:")
credit_slider = widgets.IntSlider(value=650, min=300, max=850, step=10, description="Credit Score:")
employment_status = widgets.Dropdown(options=['Employed', 'Unemployed'], value='Employed', description='Employment:')


output = widgets.Output()


def check_loan_eligibility(income, credit_score, employment):
    if credit_score >= 750 and income >= 50000 and employment == 'Employed':
       print("loan status:Approved, 5% interest rate")
    elif credit_score >= 650 and income >= 50000 and employment == 'Employed':
          print("loan status:Approved, 8% interest rate")
    else:
          print("loan status:Not Approved:(")

interact(check_loan_eligibility,
         income=income_slider,
         credit_score=credit_slider,
         employment=employment_status)
```

## **🟢 Part 2: Investment Risk Assessment [4 marks]**  

📌 **Requirement:**  
SecureBank offers **investment analysis services**. The system should evaluate a **portfolio of stocks** and classify the risk level based on stock returns:  
✔ If **any stock has a negative return**, the portfolio is **High Risk**.  
✔ If **all stocks have positive returns**, but at least one return is below 5%, classify as **Medium Risk**.  
✔ If **all stock returns are 5% or above**, classify as **Low Risk**.  

### **Implementation Steps:**  
🔹 **Step 1:** Use loops to iterate through the list of stock returns.  
🔹 **Step 2:** Implement `if-elif` conditions to classify the risk.  
🔹 **Step 3:** Write a Python program using **ipywidgets** to allow users to **input stock returns interactively** and receive a risk assessment.

---

```python
import ipywidgets as widgets
from ipywidgets import interact


def classify_risk(return_rate):
    if return_rate < 0:
        return "High Risk"
    elif return_rate < 0.05:
        return "Moderate Risk"
    else:
        return "Low Risk"

stock1_slider = widgets.FloatSlider(
    value=0.05, min=-0.2, max=0.2, step=0.01,
    description='Stock 1 Return:'
)

stock2_slider = widgets.FloatSlider(
    value=0.05, min=-0.2, max=0.2, step=0.01,
    description='Stock 2 Return:'
)

stock3_slider = widgets.FloatSlider(
    value=0.05, min=-0.2, max=0.2, step=0.01,
    description='Stock 3 Return:'
)

def assess_portfolio(ret1, ret2, ret3):
    print(f"Stock 1: {classify_risk(ret1)} (Return: {ret1*100:.1f}%)")
    print(f"Stock 2: {classify_risk(ret2)} (Return: {ret2*100:.1f}%)")
    print(f"Stock 3: {classify_risk(ret3)} (Return: {ret3*100:.1f}%)")

interact(assess_portfolio, ret1=stock1_slider, ret2=stock2_slider, ret3=stock3_slider)
```

## **🟢 Part 3: Loan Repayment Tracker [4 marks]**  

📌 **Requirement:**  
Customers who receive a loan should be able to track their **loan balance** as they make monthly payments. The system should:  
✔ Start with an **initial loan balance** (e.g., PKR 500,000).  
✔ Deduct a **fixed monthly payment** (e.g., PKR 25,000).  
✔ Continue tracking until **loan balance reaches zero**.  
✔ Display the remaining balance **after each payment**.  

### **Implementation Steps:**  
🔹 **Step 1:** Choose an appropriate loop (`for` or `while`).  
🔹 **Step 2:** Ensure the loop **stops once the loan is fully paid**.  
🔹 **Step 3:** Implement a Python program using **ipywidgets** to **simulate loan repayment interactively**.

---

```python
import ipywidgets as widgets
from IPython.display import display

loan_amount = widgets.IntText(value=500000, description="Loan Amount:")
monthly_pay = widgets.IntSlider(value=25000, min=1000, max=50000, step=1000, description="Monthly Pay:")
interest_rate = widgets.FloatSlider(value=5.0, min=0.0, max=20.0, step=0.5, description="Annual Rate (%):")
simulate_button = widgets.Button(description="Simulate Repayment", button_style='info')
output_area = widgets.Output()

display(loan_amount, interest_rate, monthly_pay, simulate_button, output_area)


def simulate_repayment(btn):
    output_area.clear_output()
    with output_area:
        balance = float(loan_amount.value)
        rate = interest_rate.value / 100
        payment = float(monthly_pay.value)
        month = 0

        if payment <= balance * rate / 12:
            print("Monthly payment is too low to cover interest. Increase payment.")
            return

        while balance > 0:
            month += 1
            balance += balance * (rate / 12)

            balance -= payment
            print(f"Month {month}: Remaining balance = PKR {balance:,.2f}")


            if month > 500:
                print("Simulation stopped (too many months).")
                break

        if balance <= 0:
            print(f"Loan fully paid in {month} months!")


simulate_button.on_click(simulate_repayment)
```


## **🟢 Part 4: Stock Price Monitoring and Trading Strategy [4 marks]**  

📌 **Requirement:**  
A stock trader wants to **track stock prices daily** and **sell when the price reaches PKR 200**. The system should:  
✔ Iterate through a **list of stock prices**.  
✔ **Skip missing stock data** (`None` values).  
✔ Stop tracking **once the price reaches PKR 200**.  

### **Implementation Steps:**  
🔹 **Step 1:** Handle missing stock data using `continue`.  
🔹 **Step 2:** Stop tracking once the stock hits the target price (`break`).  
🔹 **Step 3:** Write a Python program using **ipywidgets** to **process stock prices interactively** and trigger alerts when conditions are met.

---

```python
import random
import math
import ipywidgets as widgets
from IPython.display import display

prices = [100]
for i in range(10):
    prices.append(prices[-1] + random.uniform(-5, 5))

sell_price = 200

sell_slider = widgets.FloatSlider(value=200.0, min=50, max=300, step=1.0, description="Sell Above:")
buy_slider = widgets.FloatSlider(value=95.0, min=50, max=150, step=1.0, description="Buy Below:")
next_button = widgets.Button(description="Next Tick")
status_output = widgets.Output()

display(sell_slider, buy_slider, next_button, status_output)

price_iter = iter(prices)

def on_next_tick(b):
    with status_output:
        try:
            price = next(price_iter)
        except StopIteration:
            print("No more data.")
            return
        print(f"Price: ${price:.2f}")

        if price is None or math.isnan(price):
            print("Missing data, skipping.")
            return

        if price >= sell_slider.value:
            print("💰 Sell! Price is high.")
            next_button.disabled = True
            return

        elif price >= sell_price:
            print("Price reached PKR 200! Stopping tracking.")
            next_button.disabled = True
            return

        elif price <= buy_slider.value:
            print(" Buy more! Price is low.")
        else:
            print("Keep holding. No action needed.")

next_button.on_click(on_next_tick)
```


## **🟢 Part 5: Currency Exchange Rate Tracker [4 marks]**  

📌 **Requirement:**  
SecureBank provides **real-time currency exchange tracking**. The system should:  
✔ Start at **PKR 290/USD**.  
✔ Increase by **1 PKR per day** until it reaches **PKR 300/USD**.  
✔ Print exchange rates daily and stop when the **target rate is reached**.  

### **Implementation Steps:**  
🔹 **Step 1:** Choose a suitable loop (`for` or `while`).  
🔹 **Step 2:** Stop the loop when the exchange rate reaches the target.  
🔹 **Step 3:** Implement a Python program using **ipywidgets** to **track the currency exchange rate interactively**.

---

```python
import ipywidgets as widgets
from IPython.display import display

rate = 290
target_rate = 300
rate_per_day = 1
exchange_rates = [rate]

while rate < target_rate:
    rate += rate_per_day
    exchange_rates.append(rate)


day_slider = widgets.IntSlider(value=0, min=0, max=len(exchange_rates)-1, description="Day:")
rate_display = widgets.Output()

def show_rate(day):
    rate_display.clear_output()
    with rate_display:
        current_rate = exchange_rates[day]
        print(f"Day {day}: Exchange Rate = PKR {current_rate}/USD")
        if day > 0:
            diff = current_rate - exchange_rates[day-1]
            trend = "went up " if diff > 0 else "went down " if diff < 0 else "no change"
            print(f"Trend: {trend} (change: {diff:+.2f})")

widgets.interactive(show_rate, day=day_slider)
display(day_slider, rate_display)
```
